**Handling Categorical Data**

```python
data = pd.get_dummies(data, drop_first=True)
```

**Splitting Features and Targets**

```python
x = data.drop(columns=['diagnosis_M'])
y = data['diagnosis_M']
```

```python
# Cell: Handling Missing and Infinite Values in Features
# Replace missing values with the mean of each column
x = x.fillna(x.mean())

# Replace infinite values with large finite numbers
x = x.replace([np.inf, -np.inf], np.nan)
x = x.fillna(x.mean())
```

```python
# Additional Data Cleaning and Debugging for NaNs and Zero Variance Columns

# Check for any remaining NaN values after cleaning
print("NaN values in each column before scaling:")
print(x.isna().sum())

# Drop any columns with zero variance (constant columns)
x = x.loc[:, x.apply(pd.Series.nunique) != 1]

# Check again if any columns still contain NaN or infinite values
x.replace([np.inf, -np.inf], np.nan, inplace=True)
x.fillna(x.mean(), inplace=True)

# Confirm if there are any NaN values left
print("NaN values in each column after cleaning:")
print(x.isna().sum())
```

**Normalization**

```python
scaler = StandardScaler()
x_scaled = scaler.fit_transform(x)
x_scaled_df = pd.DataFrame(x_scaled, columns=x.columns)
```

**Outlier Detection**

```python
def detect_outliers_iqr(df):
    outliers = pd.DataFrame()
    for column in df.columns:
        if np.issubdtype(df[column].dtype, np.number):
            Q1 = df[column].quantile(0.25)
            Q3 = df[column].quantile(0.75)
            IQR = Q3 - Q1
            lower_bound = Q1 - 1.5 * IQR
            upper_bound = Q3 + 1.5 * IQR
            outliers_in_column = df[(df[column] < lower_bound) | (df[column] > upper_bound)]
            outliers = pd.concat([outliers, outliers_in_column], axis=0)
    return outliers.drop_duplicates()

outliers = detect_outliers_iqr(data)
print(f"Number of outliers detected: {len(outliers)}")
```

**Removing Outliers**

```python
outliers = outliers[outliers.index.isin(data.index)]
data = data.drop(outliers.index, axis=0)
print("Data after removing outliers:")
print(data)
```

**Balancing the Data with SMOTE**

```python
import pandas as pd
import numpy as np
from sklearn.impute import SimpleImputer
from imblearn.over_sampling import SMOTE

# Convert x_train to a DataFrame
x_train_df = pd.DataFrame(x_train)

# Drop columns that contain all NaN values
x_train_df = x_train_df.loc[:, x_train_df.notna().any(axis=0)]

# Impute missing values
imputer = SimpleImputer(strategy="mean")
x_train_imputed = imputer.fit_transform(x_train_df)  # Result is a NumPy array

# Apply SMOTE
smote = SMOTE(random_state=42)
x_train_resampled, y_train_resampled = smote.fit_resample(x_train_imputed, y_train)
```

**Visualization of Class Distribution Before and After SMOTE**

```python
plt.figure(figsize=(6, 4))
sns.countplot(x=y)
plt.title("Class Distribution Before SMOTE")
plt.xlabel("Class (0: Benign, 1: Malignant)")
plt.ylabel("Count")
plt.show()
```

```python
plt.figure(figsize=(6, 4))
sns.countplot(x=y_train_resampled)
plt.title("Class Distribution After SMOTE")
plt.xlabel("Class (0: Benign, 1: Malignant)")
plt.ylabel("Count")
plt.show()
```

**Training the AdaBoost Classifier**

```python
base_classifier = DecisionTreeClassifier(max_depth=1)
ada_boost = AdaBoostClassifier(estimator=base_classifier, n_estimators=50, algorithm='SAMME', random_state=42)
ada_boost.fit(x_train_resampled, y_train_resampled)
```

**Evaluating the Model**

```python
y_pred = ada_boost.predict(x_test)

accuracy = accuracy_score(y_test, y_pred)
precision = precision_score(y_test, y_pred)
recall = recall_score(y_test, y_pred)
f1 = f1_score(y_test, y_pred)
```

```python
print("Accuracy:", accuracy)
print("Precision:", precision)
print("Recall:", recall)
print("F1 Score:", f1)
```
